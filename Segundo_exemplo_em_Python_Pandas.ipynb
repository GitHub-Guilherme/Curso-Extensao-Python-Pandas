```python
# 
# Importando biblioteca pandas no Python

import pandas as pd
```

```python
# Ler o arquivo importado na bibliota Pandas ( as pd)

combustiveis_df = pd.read_excel("ca-2021-02.xlsx")
```

```python
print(combustiveis_df.info())
```

```python
#05
#Inserção simples de dado, com filtro head de informações para trazer apenas 3 linhas

combustiveis_df['Ativo'] = True

display(combustiveis_df.head(3))
```

```python
#06
#Criar uma coluna "Obs" que tenha nela escrito "MELHOR CIDADE" quando a coluna Municipio for igual a JOINVILLE

combustiveis_df['Obs'] = ["MELHOR CIDADE" if municipio == 'JOINVILLE' else None for municipio in combustiveis_df['Municipio']]
display(combustiveis_df.loc[combustiveis_df['Municipio'].isin(['JOINVILLE', 'INDAIATUBA', 'CAMPINAS', 'SALTO']), ['Municipio', 'Obs']])
```

```python
#07
# (por Leandro Rodrigues)
# como preencher uma coluna 'Valor de Venda - Status' que verifica o seguinte:
# Se o valor de venda for maior que 6,5 reais, ele fala que tá caro... caso contrário, está barato

import numpy as np

combustiveis_df['Valor de Venda - Status'] = np.where(combustiveis_df['Valor de Venda'] > 6.5, 'Carooooo', 'Barateza')
display(combustiveis_df[['Revenda','Municipio','Produto','Valor de Venda','Valor de Venda - Status']])

```

```python
#10
# Calcular quantos postos de gasolina por habitante temos na amostragem de combustiveis nov/2021

#Necessário baixar nova fonte de dados. 

num_habitantes_df = pd.read_csv("ibge_num_habitantes_estimado_2021.csv", sep=";")  #Antes de inserir sep=";" é importante executar o display para verificar se o separador é virgula ou ponto e virgula. Função necessária para separar em colunas
display(num_habitantes_df)
```

```python
#11
# Calcular postos de gasolina por habitante temos na amostragem de 
# combustiveis nov/2021

num_habitantes_df.rename(columns={"Estado":"Estado - Sigla"}, inplace=True)
display(num_habitantes_df)
```

```python
#12
#Faz um MERGE (CONCATENA) dois dataframes

colunas = ['Municipio', 'Estado - Sigla' ]
merge_df = combustiveis_df.merge(num_habitantes_df, how="inner", on=colunas)  #inner traz apenas valores que há número de habitantes, left traz tudo e inses NaN nas linhas que não há informações. Baseado em tabela fato
display(merge_df)
```

```python
#12B
# Faz um MERGE dos dois dataframes
colunas = ['Municipio', 'Estado - Sigla']
merge_df = combustiveis_df.merge(num_habitantes_df, how="inner", on=colunas)
display(merge_df)
print(merge_df.info())
```

```python
#Destruir coluna completamente vazia (todas linhas são nulas)
merge_df.dropna(axis='columns', inplace=True)
print(merge_df.info())
```

```python
#14
colunas=['Regiao - Sigla', 'Nome da Rua', 'Numero Rua', 
         'Bairro', 'Cep', 'Produto', 'Data da Coleta', 'Valor de Venda',
         'Unidade de Medida', 'Bandeira', 'Ativo', 'Valor de Venda - Status']
merge_df.drop(labels=colunas, axis=1, inplace=True)
print(merge_df.info())
```

```python
# Remover a linhas duplicadas
merge_df.drop_duplicates(inplace=True)
display(merge_df.head(100))
```

```python
#16
#Agrupar e contar quantos postos tem na cidade...

postos_por_municipio_df = merge_df.groupby(by=['Estado - Sigla', 'Municipio', 'NumHabitantes2021']).count()
postos_por_municipio_df.drop('CNPJ da Revenda', axis=1, inplace=True)
postos_por_municipio_df.rename(columns={"Revenda": "Número de Postos"}, inplace=True)
display(postos_por_municipio_df)
```
